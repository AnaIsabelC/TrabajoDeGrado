Genera datos aleatorios y clasifica seg√∫n los datos generados y obtenemos datos desequilibrados por nivel (pero bien clasificados)

```python
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

from sklearn.metrics import confusion_matrix

valor_maximo_temp = 37
valor_minimo_temp = 20
recorrido_temp = valor_maximo_temp - valor_minimo_temp
peso_temp = 0.48

valor_maximo_hum = 85
valor_minimo_hum = 35
recorrido_hum = valor_maximo_hum - valor_minimo_hum
peso_hum = 0.24

valor_maximo_co2 = 3000
valor_minimo_co2 = 400
recorrido_co2 = valor_maximo_co2 - valor_minimo_co2
peso_co2 = 0.16

valor_maximo_tvoc = 2500
valor_minimo_tvoc= 0
recorrido_tvoc = valor_maximo_tvoc - valor_minimo_tvoc
peso_tvco= 0.12


def asignar_puntuacion_temperatura(temperatura):
    return (valor_maximo_temp - temperatura) / recorrido_temp

def asignar_puntuacion_humedad(humedad):
    return (humedad - valor_minimo_hum) / recorrido_hum

def asignar_puntuacion_co2(co2):
    return (co2 - valor_minimo_co2) / recorrido_co2

def asignar_puntuacion_tvco(tvco):
    return (tvco - valor_minimo_tvoc) / recorrido_tvoc  # Correcci√≥n: devuelve el resultado del c√°lculo


# Las funciones y definiciones previas se mantienen igual

def generar_datos_por_nivel(cantidad_por_nivel):
    global peso_co2  # Declarar peso_co2 como variable global
    global peso_tvco

    datos_por_nivel = []

    while len(datos_por_nivel) < cantidad_por_nivel * 4:
        temperatura = random.uniform(20, 37)
        humedad = random.uniform(35, 85)
        co2 = random.randint(400, 3000)
        tvco = random.randint(0, 2500)

        peso_temperatura = asignar_puntuacion_temperatura(temperatura)*peso_temp
        peso_humedad = asignar_puntuacion_humedad(humedad)*peso_hum
        peso_co2 = asignar_puntuacion_co2(co2)*0.16
        peso_tvco = asignar_puntuacion_tvco(tvco)*0.12

        suma_pesos = peso_temperatura + peso_humedad + peso_co2 + peso_tvco
        nivel_alerta = ""

        if 0.6243122534 <= suma_pesos <= 1:
            nivel_alerta = "Muy Alto"
        elif 0.3868679095 <= suma_pesos <= 0.6208991855:
            nivel_alerta = "Alto"
        elif 0.1911466425 <= suma_pesos < 0.3834548416:
            nivel_alerta = "Medio"
        elif 0 <= suma_pesos < 0.1877335747:
            nivel_alerta = "Bajo"

        if nivel_alerta != "":
            datos_por_nivel.append({
            "Promedio Temperatura": temperatura,
            "Promedio Humedad": humedad,
            "Promedio CO2": co2,
            "Promedio TVCO": tvco,
            "Suma de Pesos": suma_pesos,
            "Nivel de Alerta General": nivel_alerta,
            })

    random.shuffle(datos_por_nivel)  # Mezclar los datos para aleatorizar su orden

    return datos_por_nivel[:cantidad_por_nivel * 4]  # Seleccionar la cantidad deseada por nivel

cantidad_por_nivel = 2500  # Cantidad deseada para cada nivel de alerta
datos_por_nivel = generar_datos_por_nivel(cantidad_por_nivel)



# Convertir los datos a un DataFrame de Pandas
df = pd.DataFrame(datos_por_nivel)

# Guardar el DataFrame como un archivo CSV
df.to_csv('datos_simulados_equilibrados_Multicriterio.csv', index=False)

# Imprimir la primera fila con las etiquetas
print(df.columns.tolist())

# Imprimir los datos sin las etiquetas
print(df.values.tolist())

# Imprimir caracter√≠sticas √∫nicas de la columna 'nivel_alerta'
print("Caracter√≠sticas de la columna 'Nivel de Alerta General':")
print(df['Nivel de Alerta General'].unique())

# Contar la cantidad de valores de cada tipo en la columna 'nivel_alerta'
conteo_nivel_alerta = df['Nivel de Alerta General'].value_counts()
print("Cantidad de valores de cada tipo en 'Nivel de Alerta General':")
print(conteo_nivel_alerta)

from google.colab import files

# files.download('datos_simulados_equilibrados_Multicriterio.csv')
```
